```python
import pandas as pd
url='https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv'
df_insurance=pd.read_csv(url)
```

```python
df_insurance.shape
```

```python
df_insurance.head()
```

```python
df_insurance['Customer Lifetime Value']=df_insurance['Customer Lifetime Value'].round(2)
```

```python
df_insurance['Total Claim Amount']=df_insurance['Total Claim Amount'].round(2)
```

```python
df_insurance.isnull().sum()
```

```python
df_insurance.dropna(subset=['Number of Open Complaints','State','Response','Months Since Last Claim','Vehicle Class', 'Vehicle Size',  'Vehicle Type'])
```

```python
df_insurance.dtypes
```

```python
df_filtered = df_insurance[(df_insurance['Total Claim Amount'] < 1000) & (df_insurance['Response'] == 'Yes')]
```

```python
df_filtered.head()
```

```python
df_yes = df_insurance[df_insurance['Response'] == 'Yes']
```

```python
summary= df_yes.groupby(['Policy Type', 'Gender'])[[ 'Customer Lifetime Value', 'Total Claim Amount']].mean()
```

```python
print(summary)
```

Corporate Auto policy for male has lowest claim amount.

```python
df_insurance['State'].value_counts()
```

```python
grouped = df_insurance.groupby(['Education', 'Gender'])
```

```python
clv_summary = grouped['Customer Lifetime Value'].agg(['max', 'min', 'median'])
```

```python
print(clv_summary)
```

Male customers with High School or Below education show the highest individual lifetime value.

Bachelor-educated females also show very high maximum values.

Doctorate holders (especially males) have comparatively lower maximum CLV.
